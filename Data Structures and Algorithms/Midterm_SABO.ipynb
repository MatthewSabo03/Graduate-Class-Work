Q1: Develop an algorithm which locates/finds the second-to-last node in a singly linked list, the last node has a next reference referring to None. (5 pts) Test the solution by creating a singly linked list object and returning the value or index of the second to last node. (5 pts) This implementation should have a linear time complexity. 

Also describe a constant time solution when using a doubly linked list. (5 pts)

To find the second to last node in a singly linked list, I will first check if the list has less than 2 nodes. If it does then I will print out an error message saying that the length of the list is less than 2. If the list is greater than 2, I will iterate through the list checking 2 nodes ahead of the current node in the list for if the 2nd node is None. If it is that means that the list is at the second to last node so then it prints the value. If it isn't, then the current node moves forward in the list and repeats until the list reaches the second to last node.

The linear Time complexity of the solution is O(n) since the code pretty much has to travel throughout the entire list to reach the solution where n is the amount of nodes in the list. 

A constant time solution for a doubly linked list is to keep a reference the tail node of the list and then call tail.prev. This would make the time complexity O(1) since it only takes 1 operation and therefore make it a constant time solution. 

```python
class Node: #abstraction of the node concept
    def __init__ (self, data):  # here we are defining what happens upon instantiation of an object (self)     
        self.item = data        # variable which passes data values
        self.next = None # next pointer var initially pointes to None which is the terminal point 

class LinkedList:
    def __init__(self): ## Initially the value of the start node is null as list will be empty when first created
        
        self.start_node = None ## We set the start node to points at the first node of the list
    
    def traverse_list(self):
        if self.start_node is None: ## First we check if the list is empty
            print("List is empty")
            return
        else:
            n = self.start_node ## we create a variable referring to the start node
            while n is not None: ## We traverse as long as n has a value, is not None
                print(n.item , " ")
                n = n.next ## We set our variable to the next pointer, once the loop reaches  the end, 'None',  it termintates
    
    def head_append(self, data):
            new_node = Node(data)
            new_node.next = self.start_node # add the reference of start_node to the next pointer variable of the new node. 
            self.start_node= new_node # As new_node is now the first node, we set the value of the start_node variable to new_node.
    
    def insert_at_index (self, index, data):
        if index == 1:
            new_node = Node(data)
            new_node.next = self.start_node
            self.start_node = new_node
        i = 1
        n = self.start_node
        while i < index-1 and n is not None:
            n = n.next
            i = i+1
        if n is None:
            print("Index out of bound")
        else: 
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
    
    def get_count(self):
        if self.start_node is None:
            return 0
        n = self.start_node
        count = 0
        while n is not None:
            count = count + 1
            n = n.next
        return count   
    
    def delete_element_by_value(self, x):
        if self.start_node is None:
            print("The list has no element to delete")
            return
        # Deleting first node 
        if self.start_node.item == x:
            self.start_node = self.start_node.ref
            return
        n = self.start_node
        while n.ref is not None:
            if n.ref.item == x:
                break
            n = n.ref
        if n.ref is None:
            print("item not found in the list")
        else:
            n.ref = n.ref.ref
            
    def find_second_to_last(self):
        currNode = self.start_node
        if currNode == None or currNode.next == None:
            print("List has less than 2 nodes")
        
        while currNode != None:
            if currNode.next.next == None:
                return currNode.item
            currNode = currNode.next

linked_list = LinkedList()
list = [1,2,3,4,5,6,7]
i=0
while i <= len(list)-1:
    linked_list.head_append((list[i]))
    i+=1

linked_list.traverse_list()
linked_list.find_second_to_last()
    
```

Q2- Develop a special case of the Stack2 Class data structure whereby the pop function always pops the second to the top node. So basically, the top node always stays the same and the node lying under it is always popped out. Also demonstrate/test your solution. (5 pts)

```python
## Start by defining a node as follows
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        
class Stack2:
    def __init__(self):
        self.top = None
        self.size = 0
                
    # Push/append operation
    def push(self, data):
        node = Node(data)  # encapsulate data in node
        if self.top: # if the stack is not empty then add point the new node's next pointer to the top element
            self.top = node
            node.next = self.top 
        else: # Otherwise/if stack is empty the enw node will become the element at the top of the stack
            self.top = node
        self.size += 1 # After pushing an item we increment the list size by 1
    
    # Pop/remove operation
    def pop(self):
        if self.top: # If list has elements, it would have one at the top of the stack
            data = self.top.data
            self.size -= 1
            if self.top.next: # if stack is not empty then we need to make next node the top node
                self.top = self.top.next
            else:
                self.top = None # Otherwise it's empty after pop operation
            return data # returns the element
        else:
            return None # if list is initially empty, we cannot perform pop therefor returns None
        
    def popSecond(self):
        old_head = self.pop()
        new_head = self.pop()
        self.push(old_head)
        return new_head

stack = Stack2()
list = [1,2,3,4,5,6,7]
i=0
while i < len(list)-1:
    stack.push((list[i]))
    i+=1

stack.popSecond()
```

Q4- Describe and develop an algorithm which swaps the positions of two contiguous/adjacent nodes in a doubly linked list and test/demonstrate the solution. (5 pts)

For this algorithm, I would first find the 2 nodes in the list and store their refrences in 2 variables so that I can keep track of their previous and next nodes. Then I would go through and swap the next refrences and previous refrences of each node and then go through and swap the next of the previous node from the group and the previous of the next node from the group 

```python
class Node:
    def __init__(self, data):
        self.item = data
        self.nref=None
        self.pref=None    

class DoublyLinkedList:
    def __init__(self):
        self.start_node = None
    
    # Inserting to an empty list
    def empty_list_insertion(self, data):
        if self.start_node is None:
            new_node = Node(data)
            self.start_node = new_node
        else:
            print("list is not empty")

    def head_insertion(self, data):
        if self.start_node is None:
            new_node = Node(data)
            self.start_node = new_node
            print("node inserted")
        else:    
            new_node = Node(data)
            new_node.nref = self.start_node
            self.start_node.pref = new_node
            self.start_node = new_node

    def tail_insertion(self, data):
        if self.start_node is None:
            new_node = Node(data)
            self.start_node = new_node
            return
        n = self.start_node
        while n.nref is not None:
            n = n.nref
        new_node = Node(data)
        n.nref = new_node
        new_node.pref = n
        
    def insert_after_item(self, x, data):
        if self.start_node is None:
            print("List is empty")
            return
        else:
            n = self.start_node
            while n is not None:
                if n.item == x:
                    break
                n = n.nref
            if n is None:
                print("node is not found")
            else:
                new_node = Node(data)
                new_node.pref = n
                new_node.nref = n.nref
                if n.nref is not None:
                    n.nref.prev = new_node
                n.nref = new_node

    def insert_before_item(self, x, data):
        if self.start_node is None:
            print("List is empty")
            return
        else:
            n = self.start_node
            while n is not None:
                if n.item == x:
                    break
                n = n.nref
            if n is None:
                print("item not in the list")
            else:
                new_node = Node(data)
                new_node.nref = n
                new_node.pref = n.pref
                if n.pref is not None:
                    n.pref.nref = new_node
                n.pref = new_node
    
    def list_traversal(self):
        if self.start_node is None: 
            print("Empty list")
            return
        else: 
            n = self.start_node
            while n is not None:
                print(n.item , " ")
                n = n.nref
    
    def delete_at_start(self):
        if self.start_node is None:
            print("There is nothing to delete here")
            return 
        if self.start_node.nref is None:
            self.start_node = None
            return
        self.start_node = self.start_node.nref
        self.start_prev = None
    
    def delete_at_end(self):
        if self.start_node is None:
            print("The list has no element to delete")
            return 
        if self.start_node.nref is None:
            self.start_node = None
            return
        n = self.start_node
        while n.nref is not None:
            n = n.nref
        n.pref.nref = None
    
    def delete_specific_value(self, x):
        if self.start_node is None:
            print("The list has no element to delete")
            return 
        if self.start_node.nref is None: # If list only contains one element
            if self.start_node.item == x:
                self.start_node = None
            else:
                print("Item not found")
            return 

        if self.start_node.item == x: ## If the value located at the head
            self.start_node = self.start_node.nref
            self.start_node.pref = None
            return

        n = self.start_node ## IF list has multiple values traverse till value is located
        while n.nref is not None:
            if n.item == x:
                break
            n = n.nref
        if n.nref is not None: # if vqlue is located in a list with multiple elements 
            n.pref.nref = n.nref # set the previous element's next node to that of the element following the one to be deleted
            n.nref.pref = n.pref # set the next element's previous pointer to the element located before the node to be deleted
        else: ## If the element is located at the ail end of the list 
            if n.item == x: 
                n.pref.nref = None
            else:
                print("Element not found")

    def list_length(self):
        length = 0
        if self.start_node is None: 
            print("List is empty")
            return
        else: 
            n = self.start_node
            while n is not None:
                n = n.nref
                length +=1
        print("Length of the Doubly Linked List: " + str(length))
        
    def _find_nodes(self, x, y):
        # private function to find nodes in list
        temp1 = None
        temp2 = None
        temp = self.start_node
        
        # traverses through list to check for nodes
        while temp != None:
            if temp.item == x:
                temp1 = temp
            elif temp.item == y:
                temp2 = temp
            temp = temp.nref
        
        return [temp1, temp2]
    
    def swap(self, node1, node2):
        # Need 2 new nodes to store values and find nodes in list
        n = self._find_nodes(node1, node2)
        n1 = n[0]
        n2 = n[1]
        
        # Checks if either node is at the start of the list and sets head accoridngly 
        if n1 is self.start_node:
            self.start_node = n2
        elif n2 is self.start_node:
            self.start_node = n1
        
        # Swaps reference of next nodes in n1 and n2
        n1Next = n1.nref
        n1.nref = n2.nref
        n2.nref = n1Next
        
        # Checks if either nodes is at the end of the list and then swaps the next nodes previous pointer to the other node
        if n1.nref != None:
            n1.nref.pref = n1
        if n2.nref != None:
            n2.nref.pref = n2
        
        # Swaps refrences of previous nodes in n1 and n2
        n1Prev = n1.pref
        n1.pref = n2.pref
        n2.pref = n1Prev
        
        # Checks in case either node is the head and then swaps next accordingly for the
        # previous node  
        if n1.pref != None:
            n1.pref.nref = n1
        if n2.pref.nref != None:
            n2.pref.nref = n2
        
        

# Test for a list with nodes
list = DoublyLinkedList()
list.empty_list_insertion(10)
for i in range(10):
    list.head_insertion(i)
print("Before swap: ")
list.list_traversal()

print()
list.swap(2,3)
print("After swap:")
list.list_traversal()
```
